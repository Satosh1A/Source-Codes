# 第6章 関係から予測しよう：回帰分析

## 21 2種類のデータの関係性の強さ：相関係数

リスト6.1

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Yu Gothic'

data = {
    "数学" : [100, 85, 90, 95, 80, 80, 75, 65, 65, 60, 55, 45, 45],
    "理科" : [94, 90, 95, 90, 85, 80, 75, 70, 60, 60, 50, 50, 48],
    "社会" : [80, 88, 70, 62, 86, 70, 79, 65, 75, 67, 75, 68, 60]
}
df = pd.DataFrame(data)
df.head()
```

リスト6.2

```python
df.plot.scatter(x="数学", y="理科", c="b")
plt.title("数学と理科の関係性")
plt.show()

df.plot.scatter(x="数学", y="社会", c="b")
plt.title("数学と社会の関係性")
plt.show()
```

### 相関係数

リスト6.3

```python
print("数学と理科 =", df.corr()["数学"]["理科"])
print("数学と社会 =", df.corr()["数学"]["社会"])
```

リスト6.4

```python
print(df.corr())
```

## 22 散布図の上に線を引いて予測

リスト6.5

```python
sns.regplot(data=df, x="数学", y="理科", line_kws={"color":"red"})
plt.show()

sns.regplot(data=df, x="数学", y="社会", line_kws={"color":"red"})
plt.show()
```

リスト6.6

```python
sns.jointplot(data=df, x="数学", y="理科", kind="reg", line_kws={"color":"red"})
plt.show()

sns.jointplot(data=df, x="数学", y="社会", kind="reg", line_kws={"color":"red"})
plt.show()
```

## 23 総当たりで表示させる散布図

###  相関行列を色の熱さで表現する：ヒートマップ

リスト6.7

```python
print(df.corr())
```

リスト6.8

```python
sns.heatmap(df.corr(), annot=True, vmax=1, vmin=-1, center=0)
plt.show()
```

### 総渡りで表示させる散布図：散布図行列

リスト6.9

```python
sns.pairplot(data=df)
plt.show()
```

リスト6.10

```python
sns.pairplot(data=df, kind="reg")
plt.show()
```

## 24 アヤメのデータを見てみよう

リスト6.11

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df = sns.load_dataset("iris")
df.head()
```

リスト6.12

```python
# 'species'列を除外したデータフレームで相関係数を計算
correlation_matrix = df.drop('species', axis=1).corr()

# 相関係数の表示
print(correlation_matrix)

```

リスト6.13

```python
sns.heatmap(correlation_matrix, annot=True, vmax=1,vmin=-1, center=0)
plt.show()
```

リスト6.14

```python
sns.pairplot(data=df)
plt.show()
```

リスト6.15

```python
df.head()
```

リスト6.16

```python
df["species"].unique()
```

リスト6.17

```python
onespecies = "setosa"

one = df[df["species"]==onespecies]
sns.heatmap(one.drop('species', axis=1).corr(), annot=True, vmax=1,vmin=-1, center=0)
plt.title(onespecies, fontsize=18)
plt.show()
```

リスト6.18

```python
onespecies = "versicolor"

one = df[df["species"]==onespecies]
sns.heatmap(one.drop('species', axis=1).corr(), annot=True, vmax=1,vmin=-1, center=0)
plt.title(onespecies, fontsize=18)
plt.show()
```
