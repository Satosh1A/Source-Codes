## 08 データのミスをチェックする：欠損値

### 欠損値の処理

リスト2.27

```python
import pandas as pd
data = {
    "国語" : [90,50,None,40],
    "数学" : [80,None,None,50]
}
idx =  ["A太","B介","C子","D郎"]
dfA = pd.DataFrame(data, index=idx)
dfA
```

リスト2.28

```python
dfA.isnull().sum()
```

リスト2.29

```python
dfB = dfA.dropna()
dfB
```

リスト2.30

```python
dfB = dfA.dropna(subset=["国語"])
dfB
```

リスト2.31

```python
dfB = dfA.fillna(dfA.mean())
dfB
```

リスト2.32

```python
dfB = dfA.fillna(method='ffill')
dfB
```
